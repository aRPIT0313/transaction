```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merged_transaction_dataset = pd.read_csv("merged_transaction_dataset.csv")
```

```python
print(merged_transaction_dataset.info())
```

```python
print(merged_transaction_dataset.isnull().sum())
merged_transaction_dataset.dropna(inplace=True)
```

```python
merged_transaction_dataset['transaction_date_time'] = pd.to_datetime(merged_transaction_dataset['transaction_date_time'], format='%d-%m-%Y %H:%M')
```

```python
print(merged_transaction_dataset.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(merged_transaction_dataset['transferred_amount'], bins=30, kde=True)
plt.title('Histogram of Transferred Amount')
plt.xlabel('Transferred Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_transaction_dataset, x='transaction_mode')
plt.title('Countplot of Transaction Mode')
plt.xlabel('Transaction Mode')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_transaction_dataset, x='transaction_date_time', y='transferred_amount')
plt.title('Time Series of Transferred Amount over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Transferred Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_transaction_dataset, x='label', y='transferred_amount')
plt.title('Boxplot of Transferred Amount by Label')
plt.xlabel('Label')
plt.ylabel('Transferred Amount')
plt.show()
```

```python
numeric_columns = merged_transaction_dataset.select_dtypes(include=['float64', 'int64']).columns
corr = merged_transaction_dataset[numeric_columns].corr()

# Visualization 6: Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_transaction_dataset, x='transaction_mode', hue='label')
plt.title('Countplot of Transaction Mode by Label')
plt.xlabel('Transaction Mode')
plt.ylabel('Count')
plt.legend(title='Label')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_transaction_dataset, x='label', y='amount_before_transaction')
plt.title('Boxplot of Amount Before Transaction by Label')
plt.xlabel('Label')
plt.ylabel('Amount Before Transaction')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_transaction_dataset, x='label', y='amount_after_transaction')
plt.title('Boxplot of Amount After Transaction by Label')
plt.xlabel('Label')
plt.ylabel('Amount After Transaction')
plt.show()
```
